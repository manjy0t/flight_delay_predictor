```python
#Add features that are lagged by 2 and 3 months to the dataset

def create_lagged_features(df, features, lags=[2, 3]):

    for feature in features:
        for lag in lags:
            df[f"{feature}_lag{lag+1}"] = df.groupby('carrier')[feature].shift(lag)

    # Drop rows with NaNs created by shifting
    df.dropna(inplace=True)
    return df
```

Let's merge the flight data with weather conditions

```python
# Load CSV data
df_wea = pd.read_csv('weather-data.csv')
df_wea['time'] = pd.to_datetime(df_wea['time'])  # or df_wea['timestamp']

df_wea.rename(columns={'temperature_2m (°C)': 'temperature_2m'}, inplace=True)
df_wea.rename(columns={'snowfall (cm)': 'snowfall'}, inplace=True)
df_wea.rename(columns={'snow_depth (m)': 'snow_depth'}, inplace=True)
df_wea.rename(columns={'cloud_cover (%)': 'cloud_cover'}, inplace=True)
df_wea.rename(columns={'wind_speed_10m (km/h)': 'wind_speed_10m'}, inplace=True)
df_wea.rename(columns={'wind_direction_10m (°)': 'wind_direction_10m'}, inplace=True)

df_wea
```

```python
import pandas as pd
import pytz  # For timezone conversions

#Target variable
# Ensure timestamps are parsed correctly (fixing string issue)
df_flights['scheduled_utc'] = pd.to_datetime(df_flights['scheduled_utc'], errors='coerce')
df_flights['runway_utc'] = pd.to_datetime(df_flights['runway_utc'], errors='coerce')

# Check for any nulls after conversion (to catch failed conversions)
if df_flights['scheduled_utc'].isna().any() or df_flights['runway_utc'].isna().any():
    print("Warning: Some datetime conversions failed. Check for invalid formats.")

# Calculate delay before rounding (use original datetime columns)
df_flights['delay'] = (df_flights['runway_utc'] - df_flights['scheduled_utc']).dt.total_seconds()

df_flights['scheduled_hour'] = pd.to_datetime(df_flights['scheduled_utc']).dt.round('H')
df_flights['runway_hour'] = pd.to_datetime(df_flights['runway_utc']).dt.round('H')

df_wea['weather_hour'] = pd.to_datetime(df_wea['time']).dt.round('H')

#Localize Weather Data to Atlanta (Eastern Time)
eastern = pytz.timezone('America/New_York')  # Atlanta timezone
df_wea['weather_hour'] = df_wea['weather_hour'].dt.tz_localize(eastern, nonexistent='shift_forward', ambiguous='NaT')

#Convert Atlanta Time to UTC
df_wea['weather_hour'] = df_wea['weather_hour'].dt.tz_convert('UTC')

merged_data = pd.merge(
    df_flights,
    df_wea,
    left_on='scheduled_hour',
    right_on='weather_hour',
    how='left'
)

merged_data.head()
merged_data.info()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Copy the dataframe to avoid modifying the original
df = merged_data.copy()

#Handle Datetime Features (Cyclical Encoding and Delay Calculation)
df['scheduled_hour'] = pd.to_datetime(df['scheduled_hour'])
df['runway_hour'] = pd.to_datetime(df['runway_hour'])

# Extract useful time components (cyclical encoding)
df['scheduled_hour_sin'] = np.sin(2 * np.pi * df['scheduled_hour'].dt.hour / 24)
df['scheduled_hour_cos'] = np.cos(2 * np.pi * df['scheduled_hour'].dt.hour / 24)
df['day_of_week'] = df['scheduled_hour'].dt.dayofweek
df['month'] = df['scheduled_hour'].dt.month

# Drop original datetime columns (optional)
df.drop(['scheduled_utc', 'revised_utc', 'runway_utc', 'scheduled_hour', 'runway_hour', 'weather_hour'], axis=1, inplace=True)

#Encode Categorical Features
categorical_cols = ['status', 'airport_iata', 'terminal', 'quality', 'aircraft_model', 'airline_iata']
binary_cols = ['is_cargo']

# Label encode binary columns
df['is_cargo'] = df['is_cargo'].astype(int)

# One-hot encode categorical columns
encoder = OneHotEncoder(drop='first', sparse_output=False)
encoded_features = pd.DataFrame(encoder.fit_transform(df[categorical_cols]), columns=encoder.get_feature_names_out(categorical_cols))
df = pd.concat([df, encoded_features], axis=1)

# Drop original categorical columns
df.drop(categorical_cols, axis=1, inplace=True)

# Scale Continuous Features
scaler = StandardScaler()
continuous_cols = ['temperature_2m', 'snowfall', 'snow_depth', 'cloud_cover', 'wind_speed_10m', 'wind_direction_10m']

# Fill missing values
imputer = SimpleImputer(strategy='mean')
df[continuous_cols] = imputer.fit_transform(df[continuous_cols])

# Scale the data
df[continuous_cols] = scaler.fit_transform(df[continuous_cols])

#Drop Redundant or Irrelevant Columns
df.drop(['flight_number', 'aircraft_reg'], axis=1, inplace=True)

df
```

One-hot enconding introduced a lot of new features, let's perform tests to see how we can reduce dimensionality.

```python
#Wew start by testing for feature importance and keeping the 150 most important features

from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
X = df.drop(['delay', 'time'], axis=1)
y = df['delay']

rf.fit(X, y)
importances = rf.feature_importances_
feature_importance_df = pd.DataFrame({'feature': X.columns, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

# Select top 150 features
top_features = feature_importance_df.head(150)['feature'].values
df = df[top_features]
```

```python
#Next, we remove features with little variance

df = pd.concat([df, y], axis=1) #Add back delay

from sklearn.feature_selection import VarianceThreshold

selector = VarianceThreshold(threshold=0.01)  # Adjust the threshold as needed
X_selected = selector.fit_transform(df.drop('delay', axis=1))
new_df = pd.DataFrame(X_selected, columns=df.drop('delay', axis=1).columns[selector.get_support()])

new_df = pd.concat([new_df, y], axis=1) #Add back delay
new_df
```

```python
#Now we also remove highly correlated features

correlation_matrix = new_df.corr().abs()
upper = correlation_matrix.where(np.triu(np.ones(correlation_matrix.shape), k=1).astype(bool))

to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]
new_df.drop(to_drop, axis=1, inplace=True)
```

```python
import optuna
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from sklearn.model_selection import train_test_split

# Define Optuna objective for fine-tuning the existing MLP model
def objective(trial, X_train, y_train, X_test, y_test):
    learning_rate = trial.suggest_loguniform('learning_rate_init', 1e-4, 1e-2)
    hidden_layer_size = trial.suggest_int('hidden_layer_size', 50, 200)

    # Update the model's hyperparameters but keep the learned weights
    mlp_model.set_params(
        hidden_layer_sizes=(hidden_layer_size,),
        learning_rate_init=learning_rate,
        max_iter=300
    )

    # Continue training the model
    mlp_model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = mlp_model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)

    return mse
```

```python
# Prepare data for new_df
X_new = new_df.drop(columns=['delay'])
y_new = new_df['delay']

X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(
    X_new, y_new, test_size=0.2, random_state=42
)

# Optimize with Optuna for new_df
study_new = optuna.create_study(direction='minimize')
study_new.optimize(lambda trial: objective(trial, X_train_new, y_train_new, X_test_new, y_test_new), n_trials=50)

print(f"Best MSE for new_df: {study_new.best_value}")
print(f"Best Hyperparameters for new_df: {study_new.best_params}")
```

```python
from sklearn.metrics import mean_squared_error, r2_score

best_params = study_new.best_params

# Rename the incorrect parameter key
if 'hidden_layer_size' in best_params:
    best_params['hidden_layer_sizes'] = best_params.pop('hidden_layer_size')

print(f"Corrected Best Hyperparameters: {best_params}")

mlp_model.set_params(**best_params)

# Retrain and Evaluate
mlp_model.fit(X_train_new, y_train_new)
y_pred_new = mlp_model.predict(X_test_new)

mse_new = mean_squared_error(y_test_new, y_pred_new)
r2_new = r2_score(y_test_new, y_pred_new)

print(f"Test MSE after fine-tuning: {mse_new}")
print(f"Test R² Score after fine-tuning: {r2_new}")
```
