## <span style='color:#ff5f27'>  Get Weather Forecast data</span>

```python
import requests
import pandas as pd
from datetime import datetime

# Coordinates for ATL Airport (Hartsfield-Jackson Atlanta International Airport)
lat = 33.6407  # Latitude for ATL
lon = -84.4279  # Longitude for ATL

# Open-Meteo API URL for hourly forecast with the requested parameters
url = (
    f"https://api.open-meteo.com/v1/forecast"
    f"?latitude={lat}&longitude={lon}"
    f"&hourly=temperature_2m,snowfall,snow_depth,weathercode,cloudcover,wind_speed_10m,wind_direction_10m"
    f"&timezone=America/New_York"
)

# Send GET request to Open-Meteo API
response = requests.get(url)

# Check the response status
if response.status_code == 200:
    weather_data = response.json()

    # Extract the hourly forecast data
    hourly_data = weather_data.get('hourly', {})

    if hourly_data:
        # Prepare the data for DataFrame conversion
        times = hourly_data.get('time', [])
        temperatures = hourly_data.get('temperature_2m', [])
        snowfall = hourly_data.get('snowfall', [])
        snow_depth = hourly_data.get('snow_depth', [])
        weather_codes = hourly_data.get('weathercode', [])
        cloud_cover = hourly_data.get('cloudcover', [])
        wind_speeds = hourly_data.get('wind_speed_10m', [])
        wind_directions = hourly_data.get('wind_direction_10m', [])

        # Create a list of dictionaries for each hour's forecast
        forecast_data = []
        for i in range(len(times)):
            # Adjust the datetime format to handle timestamps like '2025-01-07T00:00'
            time = datetime.strptime(times[i], "%Y-%m-%dT%H:%M").strftime("%Y-%m-%d %H:%M:%S")
            forecast_data.append({
                "time": time,
                "temperature_2m": temperatures[i],
                "snowfall": snowfall[i],
                "snow_depth": snow_depth[i],
                "weather_code": weather_codes[i],
                "cloud_cover": cloud_cover[i],
                "wind_speed_10m": wind_speeds[i],
                "wind_direction_10m": wind_directions[i],
            })

        # Convert the list of dictionaries to a pandas DataFrame
        weather_df = pd.DataFrame(forecast_data)

        # Print the DataFrame
        print(weather_df)

    else:
        print("No hourly weather data found in the response.")
else:
    print(f"Error: Unable to fetch data. HTTP Status Code: {response.status_code}")
    print(f"Response Text: {response.text}")  # Print the response to debug
```

```python
df['time'] = pd.to_datetime(df['time'])  # Handles milliseconds and 'T'
weather_df['time'] = pd.to_datetime(weather_df['time'])  # Handles milliseconds and 'T'
```

```python
merged_table = pd.merge(df, weather_df, on='time', how='inner')

# Display the merged table
merged_table
```

```python

merged_table.rename(columns={'temperature_2m (°C)': 'temperature_2m'}, inplace=True)
merged_table.rename(columns={'snowfall (cm)': 'snowfall'}, inplace=True)
merged_table.rename(columns={'snow_depth (m)': 'snow_depth'}, inplace=True)
merged_table.rename(columns={'cloud_cover (%)': 'cloud_cover'}, inplace=True)
merged_table.rename(columns={'wind_speed_10m (km/h)': 'wind_speed_10m'}, inplace=True)
merged_table.rename(columns={'wind_direction_10m (°)': 'wind_direction_10m'}, inplace=True)
merged_table.rename(columns={'airline.name': 'airline_name'}, inplace=True)
merged_table.rename(columns={'departure.gate': 'departure_gate'}, inplace=True)
merged_table.rename(columns={'departure.terminal': 'departure_terminal'}, inplace=True)
merged_table.rename(columns={'flight.number': 'flight_number'}, inplace=True)
```
